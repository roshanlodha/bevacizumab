```python
from lifelines.statistics import logrank_test

group = expr_sub > mean_expr
#group = expr_sub > (mean_expr + 1.5 * expr_sub.std())

surv_data = surv.loc[expr_sub.index]
T = surv_data['Overall Survival (months)']
E = surv_data['Overall Survival'].str.startswith('1:DECEASED').astype(int)

kmf = KaplanMeierFitter()

plt.figure(figsize=(8,6))

for label, mask in zip(['Above Mean', 'Below Mean'], [group, ~group]):
    # Remove NaNs from T and E for each group
    valid_idx = (~T[mask].isnull()) & (~E[mask].isnull())
    kmf.fit(T[mask][valid_idx], event_observed=E[mask][valid_idx], label=label)
    kmf.plot_survival_function(ci_show=False)

plt.title('Survival Analysis by EGR1 Expression GLASS')
plt.xlabel('Months')
plt.ylabel('Survival Probability')
plt.legend()
# Calculate log-rank test p-value between the two groups
T_high = T[group]
E_high = E[group]
T_low = T[~group]
E_low = E[~group]

# Remove NaNs
valid_high = (~T_high.isnull()) & (~E_high.isnull())
valid_low = (~T_low.isnull()) & (~E_low.isnull())

results = logrank_test(
    T_high[valid_high], T_low[valid_low],
    event_observed_A=E_high[valid_high], event_observed_B=E_low[valid_low]
)
pval = results.p_value

plt.text(
    0.6, 0.1, f'p = {pval:.4g}',
    transform=plt.gca().transAxes, fontsize=12, bbox=dict(facecolor='white', alpha=0.7)
)
plt.show()
```

```python
from scipy.stats import ttest_ind

# Define gene sets
angiogenesis_genes = [
    'VEGFA', 'FLT1', 'KDR', 'ANGPT1', 'ANGPT2', 'TEK', 'FGF2', 'PDGFB', 'PDGFRB', 'EGF', 'EGFR', 'HIF1A', 'MMP2', 'MMP9'
]
naba_collagen_genes = [
    'COL1A1', 'COL1A2', 'COL3A1', 'COL4A1', 'COL4A2', 'COL5A1', 'COL5A2', 'COL5A3', 'COL6A1', 'COL6A2', 'COL6A3', 'COL7A1', 'COL8A1', 'COL8A2', 'COL9A1', 'COL9A2', 'COL9A3', 'COL10A1', 'COL11A1', 'COL11A2', 'COL12A1', 'COL14A1', 'COL15A1', 'COL16A1', 'COL17A1', 'COL18A1', 'COL19A1', 'COL20A1', 'COL21A1', 'COL22A1', 'COL23A1', 'COL24A1', 'COL25A1', 'COL26A1', 'COL27A1', 'COL28A1'
]

# Subset expression data for the gene sets
angiogenesis_expr = expr.loc[angiogenesis_genes, group.index].dropna()
naba_collagen_expr = expr.loc[naba_collagen_genes, group.index].dropna()

# Calculate mean expression for each sample in each gene set
angiogenesis_score = angiogenesis_expr.mean(axis=0)
naba_collagen_score = naba_collagen_expr.mean(axis=0)

# Compare scores between EGR1 high (group==True) and low (group==False)

angiogenesis_high = angiogenesis_score[group]
angiogenesis_low = angiogenesis_score[~group]
naba_collagen_high = naba_collagen_score[group]
naba_collagen_low = naba_collagen_score[~group]

angiogenesis_t, angiogenesis_p = ttest_ind(angiogenesis_high, angiogenesis_low, nan_policy='omit')
naba_collagen_t, naba_collagen_p = ttest_ind(naba_collagen_high, naba_collagen_low, nan_policy='omit')

print(f"Angiogenesis geneset: mean high={angiogenesis_high.mean():.2f}, mean low={angiogenesis_low.mean():.2f}, p={angiogenesis_p:.4g}")
print(f"NABA_COLLAGEN geneset: mean high={naba_collagen_high.mean():.2f}, mean low={naba_collagen_low.mean():.2f}, p={naba_collagen_p:.4g}")
```
